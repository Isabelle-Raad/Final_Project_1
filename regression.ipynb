```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Regression: Percent Female by Continent

```python
# Open and read the data
percent_female_continent = pd.read_csv(Path('./data/percent_female_continent.csv'))
percent_female_continent.head()
```

```python
# Make a scatter plot of the percent female by year
plt.scatter(percent_female_continent.Decade, percent_female_continent.percent_female)
plt.xlabel('Decade')
plt.ylabel('Percent Female')
plt.show()
```

```python
# Let's focus on the year after 1700 by dropping the data point before 1200
percent_female_continent = percent_female_continent[percent_female_continent.Decade > 1700]
plt.figure(figsize=(6,6))
plt.scatter(percent_female_continent.Decade, percent_female_continent.percent_female)
plt.xlabel('Decade')
plt.ylabel('Percent Female')
plt.title("Percentage of Female Artists Over Time", size=15)
plt.show()
```

```python
X = percent_female_continent.Decade.values.reshape(-1, 1)
```

```python
y = percent_female_continent.percent_female
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Decade')
plt.ylabel('Percent Female')
plt.title('Percent Female Artists Over Time')
plt.savefig("images/by_continent_regression.png")
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
